```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import mplcyberpunk
```

## Passo 2 pegar as cotações históricas

```python
tickers = ["^GSPC", "^BVSP", "BRL=X", "BTC-USD"]
dados_mercado = yf.download(tickers, period = "6mo")
dados_mercado = dados_mercado["Close"]
dados_mercado
```

## Passo 3 tratar os dados

```python
dados_mercado = dados_mercado.dropna()
dados_mercado
```

```python
dados_mercado.columns = ["DOLAR", "BITCOIN", "IBOV", "SP500"]
dados_mercado
```

## Passo 4 Criar os gráficos

```python
plt.style.use("cyberpunk")
plt.plot(dados_mercado["IBOV"])
plt.title("IBOVESPA")
plt.savefig("ibov.png")
```

```python
plt.style.use("cyberpunk")
plt.plot(dados_mercado["DOLAR"])
plt.title("DOLAR")
plt.savefig("dolar.png")
```

```python
plt.style.use("cyberpunk")
plt.plot(dados_mercado["SP500"])
plt.title("SP500")
plt.savefig("sp500.png")
```

```python
plt.style.use("cyberpunk")
plt.plot(dados_mercado["BITCOIN"])
plt.title("BITCOIN")
plt.savefig("bitcoin.png")
```

## Passo 5 Calcular retornos diários

```python
retornos_diarios = dados_mercado.pct_change()
retornos_diarios
```

```python
retorno_dolar = retornos_diarios["DOLAR"].iloc[-1]
retorno_ibov  = retornos_diarios["IBOV"].iloc[-1]
retorno_sp    = retornos_diarios["SP500"].iloc[-1]
retorno_btc   = retornos_diarios["BITCOIN"].iloc[-1]
```

```python
retorno_dolar = round(retorno_dolar * 100, 2)
retorno_ibov  = round(retorno_ibov * 100, 2)
retorno_sp    = round(retorno_sp * 100, 2)
retorno_btc   = round(retorno_btc * 100, 2)
```
